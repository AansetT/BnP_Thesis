```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
timedata= pd.read_excel('TimeOptimal.xlsx', sheet_name='Sheet1')
timedata
```

```python
# Set figure size
plt.figure(figsize=(14, 7))

# Number of instances and categories
num_instances = len(timedata['Instance'])
num_categories = len(timedata.columns[1:])

# Set width of each bar, with additional separation factor
separation_factor = 0.05
bar_width = (1 - separation_factor) / (num_instances + 1)

# Generate positions for each set of bars
positions = np.arange(num_categories)

# Plot each instance
for i, instance in enumerate(timedata['Instance']):
    instance_values = timedata.loc[timedata['Instance'] == instance].values[0][1:]
    plt.bar(positions + i * (bar_width + separation_factor / num_instances), instance_values, bar_width, label=instance, color = colors[i])

# Set x-axis tick labels
plt.xticks(positions + (num_instances - 1) * (bar_width + separation_factor / num_instances) / 2, timedata.columns[1:])

# Add legend
plt.legend()

plt.title('CPU Time Across Different Solution Approaches (Optimal BnP)')
plt.xlabel('Solution Approach')
plt.ylabel('CPU Time (minutes)')

# Show plot
plt.show()
```
